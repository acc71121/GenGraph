```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', None)

df_not_cases = pd.read_csv('produced_tables/stems/not_cases.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')
df_exect_cases = pd.read_csv('produced_tables/stems/exect_cases.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')
df_partial_cases = pd.read_csv('produced_tables/stems/partial_cases.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')
df_dif_cases = pd.read_csv('produced_tables/stems/dif_cases.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')

display(df_not_cases.head(50))

affs = df_not_cases['Affix'].unique()
print(affs, len(affs))

print('In Not_Cases:')
result_not = df_not_cases[df_not_cases['Affix'].isin(affs)].groupby('Affix').head(1)
display(result_not.head(50))

print('In Exect_Cases:')
result_exect = df_exect_cases[df_exect_cases['Affix'].isin(affs)].groupby('Affix').head(1000000)
display(result_exect)

print('In Partial_Cases:')
result_partial = df_partial_cases[df_partial_cases['Affix'].isin(affs)].groupby('Affix').head(1)
display(result_partial)


print('In Dif_Cases:')
result_dif = df_dif_cases[df_dif_cases['Affix'].isin(affs)].groupby('Affix').head(1)
display(result_dif)


```
