```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

df_mlt = pd.read_csv('produced_tables/multext.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')
df_rtl = pd.read_csv('produced_tables/rtl.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')


df_rtl = df_rtl[df_rtl['para_phon'].notna()]
df_rtl = df_rtl.drop(df_rtl.loc[(df_rtl['cat'] == 'V') & (df_rtl['stem_space'].isna())].index)
stem_sp = df_rtl.loc[df_rtl['cat'] == 'V']


stem_sp = stem_sp['stem_space'].str.split(';',expand=True).astype(str)

stem_sp.to_csv('produced_tables/stem_space/initial_stem_space_extracted.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)


print(stem_sp.shape)

display(stem_sp)

def contains_none(cell): 
    if 'None' in cell:                                                     # Function to check if cell contains 'None'
        return cell

stem_sp_contains_none = stem_sp.applymap(contains_none)

display(stem_sp_contains_none[stem_sp_contains_none.any(axis=1)])
stem_sp_contains_none[stem_sp_contains_none.any(axis=1)].to_csv('produced_tables/stem_space/initial_stem_space_none_in_values.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)

def remove_none(cell):
    cell = cell.replace(",None,", ",")
    cell = cell.replace("None,", "")
    cell = cell.replace(",None", "")
    cell = cell.replace("None", "")
    return cell

stem_sp = stem_sp.applymap(remove_none)

def contains_def(cell):
    if '#DEF' in cell:
        return cell

stem_sp_contains_def = stem_sp.applymap(contains_def)
display(stem_sp_contains_def[stem_sp_contains_def.any(axis=1)])
stem_sp_contains_def[stem_sp_contains_def.any(axis=1)].to_csv('produced_tables/stem_space/initial_stem_space_#def_in_values.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)

def remove_def(cell):
    cell = cell.replace("#DEF,", "")
    cell = cell.replace(",#DEF", "")
    cell = cell.replace("#DEF", "")
    return cell

stem_sp_def = stem_sp.applymap(contains_def)
display(stem_sp_def[stem_sp_def.any(axis=1)])



stem_sp = stem_sp.applymap(remove_def)
stem_sp.to_csv('produced_tables/stem_space/cleaned_stem_space_test.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)




def contains_comma(cell):
    if ',' in cell:
        return cell


stem_sp_contains_comma = stem_sp.applymap(contains_comma)

display(stem_sp_contains_comma[stem_sp_contains_comma.any(axis=1)])
stem_sp_contains_comma[stem_sp_contains_comma.any(axis=1)].to_csv('produced_tables/stem_space/cleaned_stem_space_comma_in_values.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)

stem_sp_contains_comma = stem_sp_contains_comma[stem_sp_contains_comma.any(axis=1)]
stem_sp_contains_comma= stem_sp_contains_comma.applymap(lambda x: list(str(x).split(',')))
display(stem_sp_contains_comma)


stem_sp_contains_comma = stem_sp_contains_comma.applymap(lambda lst: [x for x in lst if x != ''])
display(stem_sp_contains_comma)
stem_sp_contains_comma.to_csv('produced_tables/stem_space/test.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)

stem_sp_contains_comma = stem_sp_contains_comma.applymap(lambda lst: '|'.join([item for item in lst if item != '']))
display(stem_sp_contains_comma)


def update_ommit_none(df1, df2):
    for row_idx, row in df2.iterrows():
        for col_idx, value in row.iteritems():
            if value != 'None':
                df1.at[row_idx, col_idx] = value

update_ommit_none(stem_sp, stem_sp_contains_comma)
stem_sp.to_csv('produced_tables/stem_space/cleaned_stem_space_test_after_updating.csv', sep='\t', lineterminator='\n', encoding='utf8', index=True)

stem_sp = stem_sp.applymap(remove_none)

stem_sp = stem_sp.apply(lambda row: df_mlt['Tag'] + row, axis=1)
stem_sp['stem_space'] = stem_sp.apply(lambda x: ';'.join(map(str, x)), axis=1)

df_rtl['stem_space'].update(stem_sp['stem_space'])
df_rtl.to_csv('produced_tables/rtl.csv', sep='\t', lineterminator='\n', encoding='utf8', index=False)
```
