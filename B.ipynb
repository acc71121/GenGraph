```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', None)
df_rtr = pd.read_csv('produced_tables/rtr.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')
df_aff = pd.read_csv('produced_tables/affixes/affixes.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')

cstr = df_aff['cstr'].to_list()
aff = df_aff['phon_affix'].to_list()

df_rtr['phon_aff_1'] = df_rtr['cstr_1'].apply(lambda x: aff[cstr.index(x)] if x in cstr else np.nan)
df_rtr['phon_aff_2'] = df_rtr['cstr_2'].apply(lambda x: aff[cstr.index(x)] if x in cstr else np.nan)

df_rtr.to_csv('produced_tables/rtr.csv', sep='\t', lineterminator='\n', encoding='utf8', index=False)
display(df_rtr.loc[df_rtr['cstr_1'] != 'X'].head(100))
```

```python
df_rtl = pd.read_csv('produced_tables/rtl.csv', sep='\t', lineterminator='\n', encoding = 'utf8', dtype='unicode')

def assign_para_desyll(row):
    if row['cat'] == 'V':
        return row['stem_space']
    else:
        return row['para_phon'].replace('.', '')

df_rtl['para_desyll'] = df_rtl.apply(assign_para_desyll, axis=1)


display(df_rtl.loc[df_rtl['cat'] == 'V'])
display(df_rtl.head(33))

df_rtl.to_csv('produced_tables/rtl.csv', sep='\t', lineterminator='\n', encoding='utf8', index=False)
```
